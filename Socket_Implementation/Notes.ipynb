# **Intro to Sockets**

Initially developed for Unix. `WinSock` is Windows implementation of Sockets.

WSA(Windows Socket API) is the API for interacting with the Windows Socekt. `ws2_32.dll` (C:\WINDOWS\System32). Here, `dll` is known as `Dynamic Link Library`.

Winsocket 2 extends Winsock 1.1 support for several protocols

## **OSI Model**

**LoopBack**

Address that points to the current machine. `localhost`

**Ports**

16 bit address, can be any integer between 1 and 65535

Ports `1 to 1024` are descibed as well knows ports and are reserved for specific applications.

It is recommended to choose a number over 1024 for testing any application.

## **What are Socket?**

A pipe between two computers on a network through which data flows.

Sockets are always two way, data can be both sent and received on a socket.

There are two types of Common Sockets i.e.

`Streaming Socket(SOCK_STREAM)`: TCP

`Datagram_Socket(SOCK_DGRAM)`: UDP

## **Working of Socket**

When we create a server a new socket it created but it is yet to be bound to an `IP` address or port number and it has the `Status: Unbound`

Then we will `bind` the server to a valid IP address and port number. It has the `Status: Bound`

After the socket is bound, it is now put into the listening state, waiting for incoming connections on the port. `Status: Listening`.

Now, if any `client` wants to communicate with the `server` then first on the clinet side a new `socket` is created which is bound to it's IP. Once the server gets the request it accpets for the communication and establishes the connection and also duplicates that socket i.e. the original socket is again ready to `listen` other request where as the duplicated socket is connected.

## **Server Functions**

1. Initialize WSA - WSAStartup()
2. Create a Socket - socket()
3. Bind the socket - bind()
4. Listen on the socket - listen()
5. Accept a connection - accept(), connect()
6. Send and Receice data - recv(), send(), recovforms(), sendto()
7. Disconnet - closesocket()

## **Client Functions**

1. Initialize WSA - WSAStartup()
2. Create a Socket - socket()
3. Connect to the server - connect()
4. Send and Receice data - recv(), send(), recovforms(), sendto()
5. Disconnet - closesocket()

## **Server Code**

The server must load the `DLL` by invoking `WSAStartup`.

It binds the server's ip address to the socket

Then listens for clients trying to establish connections

On a client connecting the server creates a new socket to handle the client server communication.
